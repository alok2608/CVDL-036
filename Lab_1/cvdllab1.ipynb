```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('cat.jpg')
plt.imshow(image)
plt.show()
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image)
plt.show()
```

```python
_, binary_image = cv2.threshold(gray_image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.axis('off') 
plt.show()
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

```python

blue_channel, green_channel, red_channel = cv2.split(image)
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(blue_channel, cmap='gray')
plt.title('Blue Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='gray')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(red_channel, cmap='gray')
plt.title('Red Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python

gaussian_blur = cv2.GaussianBlur(image, (5, 5), 0)

median_blur = cv2.medianBlur(image, 5)

average_blur = cv2.blur(image, (5, 5))

titles = ['Original Image', 'Gaussian Blur', 'Median Blur', 'Average Blur']
images = [image, gaussian_blur, median_blur, average_blur]

plt.figure(figsize=(10, 7))
for i in range(len(images)):
    plt.subplot(2, 2, i+1)
    plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)) 
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```

```python

kernel_sharpening = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])
sharpened = cv2.filter2D(img_rgb, -1, kernel_sharpening)

titles = ['Original Image', 'Sharpened Image']
images = [img_rgb, sharpened]


plt.figure(figsize=(10, 7))
for i in range(len(images)):
    plt.subplot(2, 2, i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
